```python
dx.head()
```

```python
apnea_rows = dx[dx['DX_NAME'].str.contains('apnea', case=False, na=False)]
apnea_count = apnea_rows.shape[0]
apnea_count
```

```python
distinct_apnea_rows = apnea_rows.drop_duplicates(subset=['STUDY_PAT_ID'])
```

```python
merged = pd.read_csv('/Users/lisonghe/Library/CloudStorage/OneDrive-JohnsHopkins/Semester 2/602 Research and Design in Data Mining/merged.csv')
merged1 = merged.merge(distinct_apnea_rows[['STUDY_PAT_ID', 'DX_NAME']], on='STUDY_PAT_ID', how='left')
merged1
```

```python
merged1['HAVE_APNEA?'] = np.where(merged1['DX_NAME'].isna(), 'No', 'Yes')
merged1.rename(columns={'DX_NAME': 'APNEA_NAME'}, inplace=True)
merged1.head(10)
```

```python
merged1.to_csv('merged.csv', index=False)
```

```python
condition1 = (merged1['AHI'] != 0) 
condition2 = (merged1['HAVE_APNEA?'] == 'Yes')  
AHI_accuracy = (condition1 & condition2).sum() / merged1.shape[0]
AHI_accuracy
```
