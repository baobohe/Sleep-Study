```python
import matplotlib.pyplot as plt
filtered_data = merged.dropna(subset=['Power alpha_variance', 'HAVE_APNEA?'])
sizes = 2
for condition, color in zip([True, False], ['red', 'blue']):
    condition_data = filtered_data[filtered_data['HAVE_APNEA?'] == condition]
    plt.scatter(condition_data['Power alpha_variance'], condition_data.index, 
                c=color, s=sizes, label=f'Apnea: {"Yes" if condition else "No"}')
plt.title('Scatter Plot of Alpha Power Variance')
plt.xlabel('Alpha Power Variance')
plt.ylabel('Index')
plt.legend(title='Condition')
plt.show()
```

```python
import matplotlib.pyplot as plt
filtered_data = merged.dropna(subset=['Power theta_variance', 'HAVE_APNEA?'])
sizes = 2
for condition, color in zip([True, False], ['red', 'blue']):
    condition_data = filtered_data[filtered_data['HAVE_APNEA?'] == condition]
    plt.scatter(condition_data['Power theta_variance'], condition_data.index, 
                c=color, s=sizes, label=f'Apnea: {"Yes" if condition else "No"}')
plt.title('Scatter Plot of theta Power Variance')
plt.xlabel('Theta Power Variance')
plt.ylabel('Index')
plt.legend(title='Condition')
plt.show()
```

```python
import matplotlib.pyplot as plt
filtered_data1 = merged.dropna(subset=['Power alpha_variance', 'HAVE_APNEA?'])
filtered_data2 = merged.dropna(subset=['Power theta_variance', 'HAVE_APNEA?'])
sizes = 2
for condition, color in zip([True, False], ['red', 'blue']):
    condition_data1 = filtered_data1[filtered_data1['HAVE_APNEA?'] == condition]
    condition_data2 = filtered_data2[filtered_data2['HAVE_APNEA?'] == condition]
    plt.scatter(condition_data1['Power alpha_variance'], condition_data2['Power theta_variance'], 
                c=color, s=sizes, label=f'Apnea: {"Yes" if condition else "No"}')
plt.title('Scatter Plot of Alpha vs Theta Power Variance')
plt.xlabel('Alpha Power Variance')
plt.ylabel('Theta Power Variance')
plt.legend(title='Condition')
plt.show()
```
