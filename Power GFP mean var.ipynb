```python
import os
import pandas as pd

# Given paths (replace these with the actual paths on your machine)
power_folder_path = '/Users/lisonghe/Library/CloudStorage/OneDrive-JohnsHopkins/Semester 2/602 Research and Design in Data Mining/sleep data/Data Mining/Power'  # This should be replaced with the actual path
merged_file_path = '/Users/lisonghe/Library/CloudStorage/OneDrive-JohnsHopkins/Semester 2/602 Research and Design in Data Mining/merged.csv'

# Read the merged.csv file
merged_df = pd.read_csv(merged_file_path)
merged_df = merged_df.iloc[:, :12]
# Initialize a DataFrame to hold mean and variance data for Power files
power_data = []

# Function to process each CSV file and calculate mean and variance for each frequency band
def process_file(file_path):
    df = pd.read_csv(file_path)
    return {
        'Power delta_mean': df['delta'].mean(), 'Power theta_mean': df['theta'].mean(), 
        'Power alpha_mean': df['alpha'].mean(), 'Power sigma_mean': df['sigma'].mean(), 
        'Power beta_mean': df['beta'].mean(), 'Power delta_variance': df['delta'].var(), 
        'Power theta_variance': df['theta'].var(), 'Power alpha_variance': df['alpha'].var(),
        'Power sigma_variance': df['sigma'].var(), 'Power beta_variance': df['beta'].var()
    }

# Processing Power files and collecting the mean and variance data
for file_name in os.listdir(power_folder_path):
    if file_name.endswith('.csv'):
        # Extract the study ID from the file name
        study_id = file_name.split('_')[1]
        # Construct the full file path
        file_path = os.path.join(power_folder_path, file_name)
        # Process the file and calculate the statistics
        stats = process_file(file_path)
        # Add the stats to the list, along with the study ID
        stats['SLEEP_STUDY_ID'] = study_id
        power_data.append(stats)

# Convert the list of dictionaries to a DataFrame
power_df = pd.DataFrame(power_data)
power_df
```

```python
# Merge with the merged.csv data
merged_df['SLEEP_STUDY_ID'] = merged_df['SLEEP_STUDY_ID'].astype(str)
power_df['SLEEP_STUDY_ID'] = power_df['SLEEP_STUDY_ID'].astype(str)
final_df = pd.merge(merged_df, power_df, on='SLEEP_STUDY_ID', how='left')
final_df.head()
```
