```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/lisonghe/Library/CloudStorage/OneDrive-JohnsHopkins/Semester 2/602 Research and Design in Data Mining/sleep data/Data Mining/ODI/64_25396_ODI.csv')
df = df[df['events_per_hour'].notna()]
df['events_per_hour'].iloc[0]
```

```python
aggregated_data = {
    'SLEEP_STUDY_ID': [],
    'ODI': []}
folder_path = '/Users/lisonghe/Library/CloudStorage/OneDrive-JohnsHopkins/Semester 2/602 Research and Design in Data Mining/sleep data/Data Mining/ODI'
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        study_id = file_name.split('_')[1]
        file_path = os.path.join(folder_path, file_name)
        aggregated_data['SLEEP_STUDY_ID'].append(study_id)
        df = pd.read_csv(file_path)
        df = df[df['events_per_hour'].notna()]
        aggregated_data['ODI'].append(df['events_per_hour'].iloc[0])
```

```python
aggregated_df = pd.DataFrame(aggregated_data)
aggregated_df 
```

```python
merged_df = pd.read_csv('/Users/lisonghe/Library/CloudStorage/OneDrive-JohnsHopkins/Semester 2/602 Research and Design in Data Mining/merged.csv')
merged_df['SLEEP_STUDY_ID'] = merged_df['SLEEP_STUDY_ID'].astype(str)
aggregated_df['SLEEP_STUDY_ID'] = aggregated_df['SLEEP_STUDY_ID'].astype(str)
final_df = pd.merge(merged_df, aggregated_df, on='SLEEP_STUDY_ID', how='left')
final_df.head(10) 
```

```python
final_df.to_csv("merged.csv", index = False)
```
